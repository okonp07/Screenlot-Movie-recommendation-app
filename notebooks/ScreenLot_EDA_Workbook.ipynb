# ScreenLot EDA Workbook

This notebook explains the exploratory data analysis side of ScreenLot using the current MovieLens 32M staging pipeline and the app-ready catalog.

## Goals

- load the staged ScreenLot data bundle
- inspect the app-ready catalog
- recreate the core charts used inside the Streamlit app
- connect the EDA narrative to the recommendation product

```python
from pathlib import Path

from screenlot import eda
from screenlot.content import DEFAULT_DATA_DIR
from screenlot.data import build_catalog, dataset_metrics, load_data_bundle

bundle = load_data_bundle(DEFAULT_DATA_DIR)
catalog = build_catalog(bundle)
metrics = dataset_metrics(bundle, catalog)

metrics
```

```python
ratings_chart = eda.ratings_distribution(bundle.train)
genre_chart = eda.genre_distribution(catalog)
year_chart = eda.movies_per_year(catalog)
top_movies_chart = eda.top_rated_movies(catalog)

ratings_chart.show()
genre_chart.show()
year_chart.show()
top_movies_chart.show()
```

## Notes

The same chart functions used here are exposed directly inside the Streamlit product. The EDA layer is therefore no longer separate from the app; it is now part of the product experience and supports the recommendation narrative in real time.
